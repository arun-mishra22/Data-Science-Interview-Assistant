```python
from datasets import load_dataset

dataset = load_dataset("tatsu-lab/alpaca")
```

```python
print(dataset)
print(len(dataset["train"]))
```

```python
keywords = [
    "machine learning", "deep learning", "data science",
    "python", "statistics", "neural network", "nlp",
    "sql", "algorithm", "model", "clustering",
    "regression", "classification"
]

def filter_ds(example):
    text = (example["instruction"] + " " + example["output"]).lower()
    return any(k in text for k in keywords)

filtered_dataset = dataset["train"].filter(filter_ds)

print(len(filtered_dataset))
```

```python
from datasets import load_dataset

dataset = load_dataset("OpenAssistant/oasst1")

```

```python
dataset.values()
```

```python
keywords = [
    "data science",
    "machine learning",
    "deep learning",
    "artificial intelligence",
    "statistics",
    "probability",
    "python",
    "pandas",
    "numpy",
    "sql",
    "nlp",
    "computer vision",
    "neural network",
    "regression",
    "classification",
    "clustering",
    "feature engineering",
    "data preprocessing",
    "data analysis",
    "data visualization",
    "mlops",
    "hyperparameter",
    "gradient descent",
    "transformer",
    "llm"
]
def filter_ds(example):
    text = example["text"].lower()
    return example["role"] == "assistant" and any(k in text for k in keywords)

filtered_dataset = dataset["train"].filter(filter_ds)
print(len(filtered_dataset))
```
